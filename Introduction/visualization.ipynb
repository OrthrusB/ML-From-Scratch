## Matplotlib and Seaborn for visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basics of Matplotlib

#Line plot
x=np.linspace(0,10, 100)
y=np.sin(x)
plt.plot(x,y)
```

```python
#label and title for each data in line plot
x=np.linspace(0,10, 100)
y=np.sin(x)
plt.plot(x,y, label='sin wave', color='red')
plt.title('Line plot example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() #used for labeling ie sin wave here
plt.show()
```

```python
#bar chart
categories=['A', 'B', 'C']
values=[10, 20, 30]
plt.bar(categories, values, color=('red', 'pink' , 'gray'))
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

## Visulization with Seaborn

```python
import seaborn as sns
```

```python
# Histogram and KDE plot:

data=np.random.randn(1000)
sns.histplot(data, kde=True, color="pink")
plt.title("histogram of randn with kde")
plt.show()
```

```python
#heatmap
matrix=np.random.rand(5, 5)
matrix
sns.heatmap(matrix,annot=True, cmap="RdBu")#cmap can be different so search in google , like RdBu, viridis, twilight
plt.title("heatmap of randn")
plt.show()
```

```python
#pairplot example
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species", markers=["o","s","D"])
plt.show()
```

```python
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df, kind="kde" ,hue="species", markers=["o","s","D"])
plt.show()
```

```python
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df, kind="hist" ,hue="species", markers=["o","s","D"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#load csv
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data cleaning all data set/Ice Cream Ratings.csv")
df
```

```python
#set data as index
df.set_index("Date", inplace=True)
df
```

```python
#scatter plot using texture rating , overall rating
sns.scatterplot(data=df, x="Texture Rating", y="Overall Rating")
```

```python
#hist plot
sns.histplot(data=df, x="Texture Rating", color='pink')

```

```python
#box plot
sns.boxplot(data=df, x="Texture Rating")
```

```python
#pie chart
import matplotlib.pyplot as plt
#mean overall rating for each texture ko rating
texture_means = df.groupby('Texture Rating')['Overall Rating'].mean()
# pie chart
plt.figure(figsize=(6,6)) #make the pie chart bigger for better viewing
plt.pie(texture_means, labels=texture_means.index, autopct='%1.1f%%', startangle=120)
plt.title('Mean Overall Rating Texture Rating')
plt.show()
```
