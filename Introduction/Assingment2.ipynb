# Group by and Aggregating

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data cleaning all data set/Flavors.csv')
df
```

```python
group_by_frame= df.groupby('Base Flavor')
group_by_frame
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').min(numeric_only=True)
```

```python
df.groupby('Base Flavor').max(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count']})
```

```python
#aggregate using flavor rating and texture rating
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count'],'Texture Rating':['mean','max','count']})
```

```python
df.groupby('Base Flavor').describe()
```

# Merge, Join and Concatenate


```python
import pandas as pd
```

```python
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data cleaning all data set/LOTR 2.csv')
df2
```

```python
df1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data cleaning all data set/LOTR.csv')
df1
```

```python
#append
pd.concat([df1, df2], ignore_index=True)
```

```python
#inner join
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
#outer join
df1.merge(df2, how='outer')
```

```python
#left join
df1.merge(df2, how='left')
```

```python
#right join
df1.merge(df2, how='right')
```
