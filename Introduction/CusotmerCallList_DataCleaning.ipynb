```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/data cleaning all data set/Customer Call List.xlsx')
df
```

```python
#remove the duplicate dataset
df=df.drop_duplicates()
df

```

```python
#drop column name "not useful column"
df=df.drop(columns="Not_Useful_Column")
df





```

```python
#fix issue with Last Name
df['Last_Name']=df['Last_Name'].str.strip('/')
df['Last_Name']=df['Last_Name'].str.strip('..')
df['Last_Name']=df['Last_Name'].str.strip('')
df

df['Last_Name']=df['Last_Name'].fillna('')
df['Last_Name']=df['Last_Name'].fillna('')
df['Last_Name']=df['Last_Name'].str.replace('_','')
df

```

```python
#in column  Paying Customer replace yes with Y and
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
#replace N/a with blank
df['Paying Customer']=df['Paying Customer'].str.replace('N/a','')


df['Address']=df['Address'].str.replace('N/a','')
df
```

```python
#replace Nan and N/a with blank in column Do_Not_Contact
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')


df['Do_Not_Contact']=df['Do_Not_Contact'].fillna('')
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('N/a','')
df


```

```python
df['Phone_Number']=df['Phone_Number'].fillna('')
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('|','')
df['Phone_Number']=df['Phone_Number'].str.replace('/','')
df['Phone_Number']=df['Phone_Number'].str.replace('-','')
df
```

```python
#in column Phone_Number put the numbers in the format 3digits-3digits-rest of the number
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```
